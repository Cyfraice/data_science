```python
#pulisco le date per poterle trattare come un valore numerico
df['data_di_misurazione']= df['data_di_misurazione'].str.replace('-', '')
df['data_di_misurazione']= df['data_di_misurazione'].str.replace('T', '')
df['data_di_misurazione']= df['data_di_misurazione'].str.replace(':', '')

df['data_di_misurazione'] = pd.to_numeric(df['data_di_misurazione'])

df.sort_values("data_di_misurazione") #ordino il dataframe in base alla data

df.head(20)
```

```python
uy=df[['comune','indice_qualita','tipologia_di_area','inquinante_misurato','data_di_misurazione']]

uy=uy[(uy.indice_qualita < 80)&(uy.data_di_misurazione > 20231005000000)&(uy.inquinante_misurato =="CO")]
uy=uy[(uy.comune == "Taranto")|(uy.comune == "Bari")|(uy.comune == "Statte")&(uy.comune == "Lecce")]
uy.comune.value_counts()
uy=uy.head(20)
#len(uy)
uy.plot.bar(x='comune', y='indice_qualita', figsize=(20,5))
```
